```python
#for manipulating the dataset
import pandas as pd

#for working on the data
import numpy as np

#for data visualizations
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("india-crime.csv")
data.head()
```

```python
data.info()
```

```python
data['STATE/UT'].unique()
```

```python
#Display total number of Kidnapping and Abduction happened in Maharashtra state.
temp = data[data['STATE/UT'] == "MAHARASHTRA"]
print("Total number of Kidnapping and Abduction happened in Maharashtra:", temp['Kidnapping and Abduction'].sum())
```

```python
#Display total number of Kidnapping and Abduction happened in Maharashtra state of RAIGAD district.
temp = data[(data['STATE/UT'] == "MAHARASHTRA") & (data['DISTRICT'] == "RAIGAD")]
print("Total number of Kidnapping and Abduction happened in Maharashtra state of RAIGAD district:", temp['Kidnapping and Abduction'].sum())
```

```python
# Display the heat map of 7 column(Rape', 'Kidnapping and Abduction',  'Dowry Deaths', 'Assault on women with intent to outrage her modesty',  'Insult to modesty of Women', 'Cruelty by Husband or his Relatives',  'Importation of Girls')
temp = data[['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 'Assault on women with intent to outrage her modesty', 'Insult to modesty of Women', 'Cruelty by Husband or his Relatives', 'Importation of Girls']]

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(temp.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap')
plt.show()
```

```python
#Display the bar plot ‘Dowry Deaths’ from 2001 to 2012 in India
temp = data[data['Year'].between(2001,2012)]

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(temp['Year'], temp['Dowry Deaths'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Dowry Deaths')
plt.title('Dowry Deaths from 2001 to 2012')
plt.show()
```

```python
#Display the bar plot ‘Dowry Deaths’ from 2001 to 2012 in state of west Bengal
temp = data[(data['Year'].between(2001,2012)) & (data['STATE/UT'] == "WEST BENGAL")]

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(temp['Year'], temp['Dowry Deaths'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Dowry Deaths')
plt.title('Dowry Deaths in West Bengal from 2001 to 2012')
plt.show()
```
